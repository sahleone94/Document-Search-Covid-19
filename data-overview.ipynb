# Overview

With the current Covid-19 pandemic, the COVID-19 Open Research Dataset Challenge (CORD-19) was announced on Kaggle. The dataset contains a corpus of 200,000 scholarly articles about COVID-19, SARS-CoV-2, and other related coronaviruses. The volumne of the corpus is the main issue.  
  
This, however, is not a new problem. The articles are a form or text data. Text data can be found in anything from documents, forms and archives to tweets and facebook posts. Any of which could number in the thousands if not millions.

Techniques have been developed to help deal with these problems. This project will focus on:
1.Topic Modeling
2.Keyword Search
3.Cosine Matrix
